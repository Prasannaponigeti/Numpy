## Assignment

```python
import numpy as np
import pandas as pd
```

```python
df_csv=pd.read_csv("Indian_Cricket_Team_Dataset.csv")
```

```python
df_csv
```

## 1: Convert 'Age' column to a NumPy array.

```python
np.array(df_csv['Age'])
```

## 2: Find the average age of players.

```python
average_age=np.mean(df_csv['Age'])
print(average_age)
```

## 3: Find the oldest and youngest player age.

```python
oldest=np.max(df_csv['Age'])
youngest=np.min(df_csv['Age'])
print('oldest:',oldest)
print('youngest:',youngest)
```

## 4: Count the total number of players above 30 years old.

```python
above_age=np.sum(df_csv['Age']>30)
print(above_age)
```

## 5: Find the median batting average.

```python
batting_avg = np.array([df_csv['Batting Average']])
median_avg = np.median(batting_avg)
print(median_avg)
```

## 6: Find the standard deviation of strike rates.

```python
strike_rate=np.array([df_csv['Strike Rate']])
print(strike_rate)
```

## 7: Find the 90th percentile of batting average.

```python
batting_avg = np.array([df_csv['Batting Average']])
percentile_90 = np.percentile(batting_avg,90)
print(percentile_90)
```

## 8: Find the number of players with strike rate above 150.

```python
strike_rate=np.array([df_csv['Strike Rate']])
above_150=np.sum(strike_rate>150)
print(above_150)
```

## 9: Reshape the first 20 batting averages into a 4x5 matrix.

```python
batting_avg=np.array([df_csv['Batting Average'][:20]])
reshape = batting_avg.reshape(4,5)
print(reshape)
```

## 10: Compute row-wise sum of reshaped matrix.

```python
matrix=np.array([reshape])
row_wise=np.sum(matrix, axis=1)
print(row_wise)
```

## 11: Compute column-wise mean of reshaped matrix.

```python
matrix=np.array([reshape])
column_wise=np.mean(matrix, axis=0)
print(column_wise)
```

## 12: Transpose the reshaped matrix.

```python
matrix=np.array([reshape])
transpose=matrix.T
print(transpose)
```

## 13: Find the variance of batting average.

```python
batting_avg=np.array([df_csv['Batting Average']])
varience=np.var(batting_avg)
print(varience)
```

## 14: Stack age and matches played horizontally.

```python
Age=np.array([df_csv['Age']])
Matches=np.array([df_csv['Matches Played']])
stacked = np.hstack((Age.reshape(-1, 1), Matches.reshape(-1, 1)))
print(stacked)
```

## 15: Split the batting average array into 5 equal parts.

```python
batting_avg=np.array(df_csv['Batting Average'])
arr=np.split(batting_avg,5)
print(arr)
```

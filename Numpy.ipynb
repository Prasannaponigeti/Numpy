# 8. Concatination, Stacking, Splitting of numpy

```python
# concatenate() : This will merge the two or more numpy arrays into one array
# 1D array concatenation
arr1 = np.array([100, 200, 1000])
arr2 = np.array([0, -5, 20])
arr = np.concatenate((arr1, arr2))
print(arr)
```

```python
# 2D arrays concatenate
# axis = 1 : horizontal concatenation (column wise)

arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [500, 800, 900]])

arr_2D = np.concatenate((arr1, arr2), axis = 1)
print(arr_2D)
```

```python
# 2D arrays concatenate
# axis = 0 : vertical concatenation (row wise)

arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [500, 800, 900]])

arr_2D = np.concatenate((arr1, arr2), axis = 0)
print(arr_2D)
```

```python
# imp point : We cannot concatenate() two different dimension arrays
# a = np.array([1,2,3])
# b= np.array([[4,5,6]])
# np.concatenate((a,b), axis = 0) # either put axis = 0
```

```python
# We can concatenate below arrays row wise but not column wise

a = np.array([[1,2,3],
              [4,5,6]])  # 2,3
b= np.array([[7,8,9]])  #1,3
np.concatenate((a,b), axis = 0)
```

### Stack() function in numpy

#### Def : Stack() function is also used for merging the numpy arrays.
#### Stack() function will add one more extra dimension in final output array

```python
arr1 = np.array([100, 200, 1000])
arr2 = np.array([0, -5, 20])

# axis = 0: row wise stacking
stacked_arr = np.stack((arr1, arr2), axis = 0)
print(stacked_arr)
```

```python
print(np.concatenate((arr1, arr2)))
```

```python
# axis = 1 : column wise stacking
stacked_arr = np.stack((arr1, arr2), axis = 1)
print(stacked_arr)
```

##### Axis value tell us where to add the new dimension
##### N : number of input array
##### axis = 0: N, a1, a2,....,ak
##### axis = 1: a1,N, a2,.....ak
##### axis = 2: a1,a2,N.....ak

```python
# 2D array stacking
arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [500, 800, 900]])

# axis = 0 : row wise
stacked_2D = np.stack((arr1, arr2), axis = 0)
print(stacked_2D)
```

```python
# 2D array stacking
arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [500, 800, 900]])

# axis = 1 column wise
stacked_2D = np.stack((arr1, arr2), axis = 1)
print(stacked_2D)
```

#### vstack(): vertical stacking
- vstack() is same as vertical concatenation

```python
arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [500, 800, 900]])
result = np.vstack((arr1, arr2))
print(result)
```

```python
np.concatenate((arr1, arr2))
```

```python
# hstack() : horizontal stacking
#hstack() is same as horizontal concatenation

arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [500, 800, 900]])
result = np.hstack((arr1, arr2))
print(result)
```

```python
arr1 = np.array([[50, 20, 2],
                 [5, 8, 9]])

arr2 = np.array([[500, 2000, 200],
                 [500, 800, 900]])

arr3 = np.array([[25, 70, 500],
                 [700, 65, 55]])

result = np.hstack((arr1, arr2, arr3))
print(result)
```
